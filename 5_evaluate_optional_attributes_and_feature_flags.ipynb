```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
# Define date to conduct the analysis with (default set to today, override if required)
date=datetime.today().strftime('%Y.%m.%d')
#date='2025.03.21'
print('Data analysis of optional attributes started for date: '+date)

# Read data from file
certificationsData = pd.read_csv('certified_devices_'+date+'.csv')
uncertifiedDataNotUnique = pd.read_csv('uncertified_devices_'+date+'.csv')

# remove historical version models which wonâ€™t get certified in the future
uncertifiedData = uncertifiedDataNotUnique.drop_duplicates(subset=['vidPid'], keep='last')
```

```python
rows=[]

# calculate availability percentage of attribute vendorLandingPageURL
certified_percent_missing = certificationsData['vendorLandingPageURL'].isnull().sum() * 100 / len(certificationsData)
uncertified_percent_missing = uncertifiedData['vendorLandingPageURL'].isnull().sum() * 100 / len(uncertifiedData)
entry=['vendorLandingPageURL', 100-certified_percent_missing, 100-uncertified_percent_missing]
rows.append(entry)

# calculate availability percentage of attribute userManualUrl
certified_percent_missing = certificationsData['userManualUrl'].isnull().sum() * 100 / len(certificationsData)
uncertified_percent_missing = uncertifiedData['userManualUrl'].isnull().sum() * 100 / len(uncertifiedData)
entry=['userManualUrl', 100-certified_percent_missing, 100-uncertified_percent_missing]
rows.append(entry)

# calculate availability percentage of attribute supportUrl
certified_percent_missing = certificationsData['supportUrl'].isnull().sum() * 100 / len(certificationsData)
uncertified_percent_missing = uncertifiedData['supportUrl'].isnull().sum() * 100 / len(uncertifiedData)
entry=['supportUrl', 100-certified_percent_missing, 100-uncertified_percent_missing]
rows.append(entry)

# calculate availability percentage of attribute productUrl
certified_percent_missing = certificationsData['productUrl'].isnull().sum() * 100 / len(certificationsData)
uncertified_percent_missing = uncertifiedData['productUrl'].isnull().sum() * 100 / len(uncertifiedData)
entry=['productUrl', 100-certified_percent_missing, 100-uncertified_percent_missing]
rows.append(entry)

# calculate availability percentage of attribute lsfUrl
certified_percent_missing = certificationsData['lsfUrl'].isnull().sum() * 100 / len(certificationsData)
uncertified_percent_missing = uncertifiedData['lsfUrl'].isnull().sum() * 100 / len(uncertifiedData)
entry=['lsfUrl', 100-certified_percent_missing, 100-uncertified_percent_missing]
rows.append(entry)

# calculate availability percentage of attribute releaseNotesUrl
certified_percent_missing = certificationsData['releaseNotesUrl'].isnull().sum() * 100 / len(certificationsData)
uncertified_percent_missing = uncertifiedData['releaseNotesUrl'].isnull().sum() * 100 / len(uncertifiedData)
entry=['releaseNotesUrl', 100-certified_percent_missing, 100-uncertified_percent_missing]
rows.append(entry)

# calculate availability percentage of attribute firmwareInformation
certified_percent_missing = certificationsData['firmwareInformation'].isnull().sum() * 100 / len(certificationsData)
uncertified_percent_missing = uncertifiedData['firmwareInformation'].isnull().sum() * 100 / len(uncertifiedData)
entry=['firmwareInformation', 100-certified_percent_missing, 100-uncertified_percent_missing]
rows.append(entry)

# calculate availability percentage of attribute otaUrl
certified_percent_missing = certificationsData['otaUrl'].isnull().sum() * 100 / len(certificationsData)
uncertified_percent_missing = uncertifiedData['otaUrl'].isnull().sum() * 100 / len(uncertifiedData)
entry=['otaUrl', 100-certified_percent_missing, 100-uncertified_percent_missing]
rows.append(entry)


optional_attribute_additional_data_results = pd.DataFrame(rows, columns=['attributeName', 'presentInCertifiedDevices', 'presentInUncertifiedDevices']) 
print(optional_attribute_additional_data_results)
# Save optional attribute additional data Data to CSV file
optional_attribute_additional_data_results.to_csv('optional_attribute_additional_data_results_'+date+'.csv', index=False)


# Flags
flag_rows=[]

# calculate deviation (from default value 0) percentage of attribute discoveryCapabilitiesBitmask
certified_deviate_percent = certificationsData['discoveryCapabilitiesBitmask'].gt(0).sum() * 100 / len(certificationsData)
uncertified_deviate_percent = uncertifiedData['discoveryCapabilitiesBitmask'].gt(0).sum() * 100 / len(uncertifiedData)
flag_rows.append(['discoveryCapabilitiesBitmask', certified_deviate_percent, uncertified_deviate_percent])

# calculate deviation (from default value 0) percentage of attribute enhancedSetupFlowOptions
certified_deviate_percent = certificationsData['enhancedSetupFlowOptions'].gt(0).sum() * 100 / len(certificationsData)
uncertified_deviate_percent = uncertifiedData['enhancedSetupFlowOptions'].gt(0).sum() * 100 / len(uncertifiedData)
flag_rows.append(['enhancedSetupFlowOptions', certified_deviate_percent, uncertified_deviate_percent])

# calculate deviation (from default value 0) percentage of attribute commissioningCustomFlow
certified_deviate_percent = certificationsData['commissioningCustomFlow'].gt(0).sum() * 100 / len(certificationsData)
uncertified_deviate_percent = uncertifiedData['commissioningCustomFlow'].gt(0).sum() * 100 / len(uncertifiedData)
flag_rows.append(['commissioningCustomFlow', certified_deviate_percent, uncertified_deviate_percent])

optional_attribute_feature_flags_results = pd.DataFrame(flag_rows, columns=['attributeName', 'deviateFromDefaultInCertifiedDevices', 'deviateFromDefaultInUncertifiedDevices']) 
print(optional_attribute_feature_flags_results)

# Save optional attribute feature flag Data to CSV file
optional_attribute_feature_flags_results.to_csv('optional_attribute_feature_flags_results_'+date+'.csv', index=False)
```
